```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snb
import plotly.express as px
```

```python
cars_data = pd.read_csv("min_carsdata.csv")
```

```python
cars_data
```

```python
cars_infos = pd.read_csv('min_carsinfo2.csv')
```

```python
cars_infos
```

```python
cars_data_all = pd.concat([cars_data, cars_infos],axis=1)
```

```python
cars_data_all.drop_duplicates()
```

```python
cars_data_all.dropna(inplace=True)
```

```python
cars_data_all.reset_index(drop=True)
```

```python
cars = cars_data_all
cars.drop('Km',axis=1, inplace=True)
```

```python
cars['price'] = cars['price'].apply(lambda x: x.split('F')[0].strip().replace(' ',''))
cars['Engine'] = cars['Engine'].apply(lambda x: x[:-4])
cars['km'] = cars['km'].apply(lambda x: x.replace(' ','')[:-2])
cars['HorsePower'] = cars['HorsePower'].apply(lambda x: x[:-3])
cars['doors'] = cars['doors'].apply(lambda x: x.replace(' ajtós',''))
cars['places'] = cars['people'].apply(lambda x: x.replace(' fő',''))
cars.drop('people', axis=1, inplace=True)
```

```python
cars['price'] = pd.to_numeric(cars['price'])
cars['Engine'] = pd.to_numeric(cars['Engine'])
cars['km'] = pd.to_numeric(cars['km'])
cars['HorsePower'] = pd.to_numeric(cars['HorsePower'])
cars['doors'] = pd.to_numeric(cars['doors'])
cars['places'] = pd.to_numeric(cars['places'])
```

```python
cars.reset_index(drop=True)
```

```python
pd.get_dummies(cars.Transmission)
```

```python
cars = pd.concat([cars, pd.get_dummies(cars.Transmission)],axis=1)
```

```python
cars = pd.concat([cars, pd.get_dummies(cars.Sale)],axis=1)
cars = pd.concat([cars, pd.get_dummies(cars.Fuel_type)],axis=1)
cars = pd.concat([cars, pd.get_dummies(cars.Condition)],axis=1)
cars = pd.concat([cars, pd.get_dummies(cars.dealership_evaluation)],axis=1)
cars
```

```python
cars['Brand'] = cars.brand_mf.apply(lambda x: x.split(' ')[0])
```

```python
cars = pd.concat([cars, pd.get_dummies(cars.Brand)],axis=1)
cars
```

```python
cars['Type'] = cars.type.apply(lambda x: x.split('/')[0] if '/' in x else x)
cars['Type'] = cars.Type.apply(lambda x: x.replace('1 év garancia', '') if '1 év garancia' in x else x)
cars['Type'] = cars.Type.apply(lambda x: x.replace('MO-i gépjármű','') if 'mo-i' in x.lower() else x)
cars['year'] = cars.year.apply(lambda x: x.replace('.','-')[:-1])

cars
```

```python
from datetime import datetime

cars['year'] = cars.year.apply(lambda x: pd.to_datetime(x))
cars['Age'] = cars.year.apply(lambda x: pd.to_datetime('today').normalize()-x.normalize())
cars['Age'] = cars.Age.apply(lambda x: x.days/365.25)
cars['Age']
```

```python
cars['Location'] = cars.dealership_loc.apply(lambda x: 1 if 'kerület' in x else 0)
```

```python
cars_new = cars.drop(['brand_mf','type','dealership_name','dealership_loc', 'dealership_evaluation','year','Fuel_type',
                      'Kialakitas', 'Condition', 'Sale', 'Transmission', 'Brand', 'Type'], axis=1)
```

```python
cars_new.info()
```

```python
cars_new['Yearly_usage_in_km'] = cars.km / cars.Age
```
